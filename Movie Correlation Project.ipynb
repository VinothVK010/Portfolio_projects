```python
# Import libraries

import pandas as pd 
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the Data

df = pd.read_csv(r'C:\Users\Vinoth VK\Desktop\movies.csv')

df = df.dropna()
df.shape

#
```

```python

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types 

df.dtypes
```

```python
# changing data types to the columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# create Year correct column

df['year_correct'] = df['released'].astype(str).str.extract(pat = '([0-9]{4})')
```

```python
df.sort_values(by = ['gross'], ascending = False,inplace = False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# scatterplot between budget vs gross

plt.scatter(x = df["budget"], y= df['gross'])

plt.title("Scatterplot Btw Budget and Gross Earnings")

plt.xlabel('Budget')

plt.ylabel('Gross')

plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"},line_kws={"color": "blue"})

plt.show()
```

```python
# let's start looking for correlation

df.corr(method = 'pearson') # pearson, kendall, spearman
```

```python
# let's see the correlation map

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(data = correlation_matrix, annot= True)

plt.title("Correlation Chart")

plt.xlabel('Movie features')

plt.ylabel('Movie features')

```

```python
# let look at company

df_numeraized = df

for col_name in df_numeraized.columns:
    if df_numeraized[col_name].dtype == 'object':
        df_numeraized[col_name] = df_numeraized[col_name].astype('category') 
        df_numeraized[col_name] = df_numeraized[col_name].cat.codes
        
```

```python
# let's look at a Chart for all

correlation_matrix = df_numeraized.corr(method = 'pearson')

sns.heatmap(data = correlation_matrix, annot= True)

plt.title("Correlation Chart")

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
correlation_mat = df_numeraized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
